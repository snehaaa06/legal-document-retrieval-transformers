# 📊 COLIEE Dataset Exploration
## Understanding the Legal Case Retrieval Data

**Purpose:** Analyze dataset statistics and characteristics


```
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

import os
import json
import statistics
from collections import Counter
import matplotlib.pyplot as plt

# Paths
BASE_DIR = "/content/drive/MyDrive/COLIEE_Dataset"
TRAIN_DOCS_DIR = f"{BASE_DIR}/train_docs"
TEST_DOCS_DIR = f"{BASE_DIR}/test_docs"
TRAIN_LABELS = f"{BASE_DIR}/labels/task1_train_labels_2025.json"
TEST_LABELS = f"{BASE_DIR}/labels/task1_test_labels_2025.json"
```

## 1️⃣ Load Labels

```
with open(TRAIN_LABELS, 'r') as f:
    train_labels = json.load(f)

with open(TEST_LABELS, 'r') as f:
    test_labels = json.load(f)

print(f"Train queries: {len(train_labels)}")
print(f"Test queries: {len(test_labels)}")
```

## 2️⃣ Analyze Label Distribution

```
train_rel_counts = [len(docs) for docs in train_labels.values()]

print("Training Set Statistics:")
print(f"  Min relevant docs: {min(train_rel_counts)}")
print(f"  Max relevant docs: {max(train_rel_counts)}")
print(f"  Mean: {statistics.mean(train_rel_counts):.2f}")
print(f"  Median: {statistics.median(train_rel_counts):.0f}")

# Visualize distribution
plt.figure(figsize=(10, 6))
plt.hist(train_rel_counts, bins=30, edgecolor='black')
plt.xlabel('Number of Relevant Documents')
plt.ylabel('Number of Queries')
plt.title('Distribution of Relevant Documents per Query')
plt.grid(alpha=0.3)
plt.show()
```

## 3️⃣ Analyze Document Lengths

```
# Sample 100 documents to analyze length
import random

doc_files = os.listdir(TRAIN_DOCS_DIR)[:100]
doc_lengths = []

for doc_file in doc_files:
    with open(os.path.join(TRAIN_DOCS_DIR, doc_file), 'r', encoding='utf-8', errors='ignore') as f:
        text = f.read()
        doc_lengths.append(len(text.split()))

print("Document Length Statistics (in words):")
print(f"  Min: {min(doc_lengths):,}")
print(f"  Max: {max(doc_lengths):,}")
print(f"  Mean: {statistics.mean(doc_lengths):,.0f}")
print(f"  Median: {statistics.median(doc_lengths):,.0f}")

# Visualize
plt.figure(figsize=(10, 6))
plt.hist(doc_lengths, bins=30, edgecolor='black')
plt.xlabel('Document Length (words)')
plt.ylabel('Frequency')
plt.title('Distribution of Document Lengths')
plt.grid(alpha=0.3)
plt.show()
```

## 4️⃣ Show Sample Documents

```
# Show a sample query and its relevant docs
sample_query = list(train_labels.keys())[0]
relevant_docs = train_labels[sample_query]

print(f"Sample Query: {sample_query}")
print(f"Relevant documents: {relevant_docs}\n")

# Load and show query text
with open(os.path.join(TRAIN_DOCS_DIR, sample_query), 'r', encoding='utf-8', errors='ignore') as f:
    query_text = f.read()

print("Query text (first 500 chars):")
print("-" * 80)
print(query_text[:500])
print("-" * 80)
```
